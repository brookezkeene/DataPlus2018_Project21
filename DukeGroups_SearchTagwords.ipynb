```python
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.chrome.options import Options
from selenium.common.exceptions import NoSuchElementException
from time import sleep

def expand_website():
    # Access DukeGroups and make soup
    while True:
        try:
           # poll the link with an arbitrary call
            bLoadMore = driver.find_element_by_xpath('//*[@id="react-app"]/div/div[2]/div/div/div[2]/div/div[2]/div[2]/button');
            bLoadMore.click()
            sleep(1);
        except NoSuchElementException:
            print("You've reached the end of the list!")
            return False
        
# Keys: tagwords
# Values: search results
match = {}

for i in tagwords:
        match[i] = [];
        options = Options()
        options.add_argument('--headless')
        options.add_argument('--disable-gpu')  # Last I checked this was necessary.
        driver = webdriver.Chrome(executable_path='/Users/derekliu/Downloads/chromedriver', chrome_options=options)
        if ' ' in i:
            j = ' '.split(i)
            link = 'https://dukegroups.com/organizations?query=' + i[0] + '%20' + i[1]
        link = 'https://dukegroups.com/organizations?query=' + i;
        driver.get(link)
        for x in range (0, 20):
            if expand_website() == False:
                break
            expand_website()
        org_name_soup = BeautifulSoup(driver.page_source, 'lxml');
        for link in org_name_soup.find_all('a'):
            temp = link.find('p', attrs={'class': "DescriptionExcerpt"})
            if temp == None:
                continue
            name = temp.previous_sibling.get_text()
            name = name[1:-1]
            match[i].append(name)
```

```python
# Keys: co-curricular names
# Values: Relevant tags
actDic = {}

for tag in tagwords:
    for act in match[tag]:
        if act in actDic:
            actDic[act].append(tag)
        else:
            actDic[act] = [tag]
```

```python
# Keys: co-curricular names
# Values: 1 or 0 for tags
actTable = {}
count = False
for act in actDic.keys():
    for allTag in tagwords:
        for myTag in actDic[act]:
                if myTag == allTag:
                    count = True
                    break
                else:  
                    count = False           
        if (act in actTable) & (count == True):
            actTable[act].append(1)
        if (act in actTable) & (count == False):
            actTable[act].append(0)
        if (act not in actTable) & (count == True):
            actTable[act] = [1]
        if (act not in actTable) & (count == False):
            actTable[act] = [0]
```
